```python
from pyspots.data import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
norm_spots = data.spots90FQ2_normalized_spot
```

```python
df = pd.DataFrame({'spot-%s' % spot.idx: spot.data for spot in norm_spots}, index=norm_spots[0].times)
```

```python
window_size = 3
# compute the rolling mean
ma = pd.rolling_mean(df, window=window_size)
# the `window_size-1` data points will be NaN, drop them
ma = ma[window_size-1:]
```

```python
min_avg = ma.head(5).mean()
max_avg = ma.tail(5).mean()
```

```python
normed = (ma - min_avg) / (max_avg - min_avg)
```

```python
normed.iloc[:,range(5)].plot()
```
